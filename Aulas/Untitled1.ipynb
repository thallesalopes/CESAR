```python
dado = input()

if not '0' in dado:
  if dado.count(" ")==1:
      entrada = dado.split(" ")
      if (int(entrada[0]) != 0) and (int(entrada[1]) != 0):
          print(int(str(int(entrada[0])+int(entrada[1])).replace('0','')))
  elif dado.count(" ")==2:
      entrada = dado.split(" ")
      n1 = int(entrada[0])
      n2 = int(entrada[1])*1000+int(entrada[2])
      print(int(str(n1+n2).replace('0','')))
  elif dado.count(" ")==3:
      entrada = dado.split(" ")
      n1 = int(entrada[0])*1000+int(entrada[1])
      n2 = int(entrada[2])*1000+int(entrada[3])
      print(int(str(n1+n2).replace('0','')))
  elif dado.count(" ")==4:
      entrada = dado.split(" ")
      n1 = int(entrada[0])*1000+int(entrada[1])
      n2 = int(entrada[2])*1000000+int(entrada[3])*1000+int(entrada[4])
      print(int(str(n1+n2).replace('0','')))
  elif dado.count(" ")==5:
      entrada = dado.split(" ")
      n1 = int(entrada[0])*1000000+int(entrada[1])*1000+int(entrada[2])
      n2 = int(entrada[3])*1000000+int(entrada[4])*1000+int(entrada[5])
      print(int(str(n1+n2).replace('0','')))
    
    
```
