```python
def fat(n):
  fatorial=1
  for i in range(n,1,-1):
    fatorial *= i 
  return fatorial  

print( fat(1) )
print( fat(5) ) 
print( fat(6))

```

```python
def fat(n):
  if type(n)!=int or n<0 :
    return "erro"
  elif n==1 or n==0: 
    return 1
  else: 
    return n*fat(n-1) 

print(fat(-1))
print(fat('a'))
print(fat(6))
```

```python
dic1 = { 1:10 , 2:20 }
dic2 = { 3:30 , 4:40 }
dic3 = { 5:50 ,6:60 }

def funcao(*dicionarios):
  dic_res ={}
  for dicionario in dicionarios:
    for chave in dicionario.keys():
      dic_res[chave]=dicionario[chave]
  return dic_res

funcao(dic1,dic2,dic3)
```

```python
import random

sorteados =[]
for i in range(10):
  aleatorio = random.randint(0,100)
  sorteados.append(aleatorio)

print("o maior número sorteado foi: ", max(sorteados))
print("o menor número sorteado foi: ", min(sorteados))
print("a média dos números sorteados foi: ", sum(sorteados)/10)
```

```python
dados = open("alunos.csv", 'r')
dados.readline()
RA_alunos =[]
media=[]
for aluno in dados:
  varRA,varAluno,varFreq,varN1,varN2,varN3,varN4 = aluno.replace('\n','').split(',')
  RA_alunos.append(varRA)
  media.append((float(varN1)+float(varN2)+float(varN3)+float(varN4))/4)
dados.close()

resultado = open("media.csv",'w')
resultado.write("RA,Média\n")
for i in range(len(RA_alunos)):
  resultado.write(RA_alunos[i]+','+str(media[i])+'\n')
resultado.close()
```
